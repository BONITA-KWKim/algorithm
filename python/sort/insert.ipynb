```python
def insert_sort(arr):
    size = len(arr)
    if 0 >= size: return None
    
    for i in range(1, size):
        picked = arr[i]
        
        if arr[i-1] < picked:
            continue
        
        for j in range(0,i):
            if picked < arr[j]:
                arr[j+1:i+1] = arr[j:i]
                arr[j] = picked
                break;
    return arr
```

```python
test_arr = [5, 1, 6, 4, 2, 3]
print(insert_sort(test_arr))
test_arr = []
print(insert_sort(test_arr))
test_arr = [3, 3, 3]
print(insert_sort(test_arr))
test_arr = [5, 1, 4, 23, 67, 34786, 222, 333, 222]
print(insert_sort(test_arr))
```

```python
import random

max_no = 1000
test_arr = []
for i in range(max_no):
    test_arr.append(random.randrange(0, 999999))

print(insert_sort(test_arr))
```
